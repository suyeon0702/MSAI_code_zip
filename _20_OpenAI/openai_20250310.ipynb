```python
from dotenv import load_dotenv
import os

load_dotenv("./.env")  # .env 파일 로드

api_key = os.getenv("AZURE_OPENAI_KEY")
endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
endpoint = endpoint[:len(endpoint)-1] # 마지막 / 제거
```

```python
import requests

# method : POST
headers = {
    "Content-Type":"application/json",
    "api-key":api_key
}

body = {
    "messages": [
        {
            "role": "system",
            "content": [
                {
                    "type": "text",
                    "text": "너는 인플루언서야. 너의 팬들 이름은 곰돌이야. 너는 라이브 중에 팬들한테 질문을 받고 있어. 광고인데 광고인 티가 나지 않게 추천해줘야 해. 약간 과시하지 않는 척 과시하는 듯한 어투를 가지고 있어."
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "오늘 점심 메뉴 추천해주세요!"
                }
            ]
        },
        {
            "role": "assistant",
            "content": [
                {
                    "type": "text",
                    "text": "[메뉴 이름]\n---\n[이유]\n---\n[추천 가게]\n---\n[주소]\n---"
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "오늘 점메추해주세요 언닝~~"
                }
            ]
        }
        
    ],
    "temperature": 1,
    "top_p": 1,
    "max_tokens": 800
}

response = requests.post(endpoint, headers=headers, json=body)
response_json = response.json()
message = response_json['choices'][0]['message']
role = message['role']
content = message['content']

print(message['content'], message['role'])

# 호출하면 된다.
```

```python
def request_gpt(prompt):
    # method : POST
    headers = {
        "Content-Type":"application/json",
        "api-key":api_key
    }

    body = {
        "messages": [
            {
                "role": "system",
                "content": [
                    {
                        "type": "text",
                        "text": "너는 인플루언서야. 너의 팬들 이름은 곰돌이야. 너는 라이브 중에 팬들한테 질문을 받고 있어. 광고인데 광고인 티가 나지 않게 추천해줘야 해. 약간 과시하지 않는 척 과시하는 듯한 어투를 가지고 있어."
                    }
                ]
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": "오늘 점심 메뉴 추천해주세요!"
                    }
                ]
            },
            {
                "role": "assistant",
                "content": [
                    {
                        "type": "text",
                        "text": "[메뉴 이름]\n---\n[이유]\n---\n[추천 가게]\n---\n[주소]\n---"
                    }
                ]
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt
                    }
                ]
            }
            
        ],
        "temperature": 1,
        "top_p": 1,
        "max_tokens": 800
    }

    response = requests.post(endpoint, headers=headers, json=body)
    response_json = response.json()
    message = response_json['choices'][0]['message']
    role = message['role']
    content = message['content']

    # print(message['content'], message['role'])
    return content

# "오늘 점메추해주세요 언닝~~"
# print(request_gpt("오늘 점메추해주세요 언닝~~"))

from IPython.display import Markdown
display(Markdown(request_gpt("오늘 저메추해주세요 언닝~~")))
```
